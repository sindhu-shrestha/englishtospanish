```python
from transformers import MarianMTModel, MarianTokenizer

# Loading model and tokenizer
model_name = 'Helsinki-NLP/opus-mt-en-es'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

```

```python
#function for translation

def interactive_translation():
    while True:
        sentence = input("Enter an English sentence to translate into Spanish: ")
        inputs = tokenizer(sentence, return_tensors="pt", padding=True)
        translated = model.generate(**inputs)
        spanish = tokenizer.decode(translated[0], skip_special_tokens=True)
        print(f"→ Spanish Translation: {spanish}")

        again = input("Do you want to translate another sentence? (yes/no): ").strip().lower()
        if again not in ['yes', 'y']:
            print("Goodbye!")
            break
```

```python
interactive_translation()
```

```python
from transformers import MarianMTModel, MarianTokenizer

# Load English to Spanish model
model_name = 'Helsinki-NLP/opus-mt-en-es'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)


```

```python
# Translate function
def translate(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", padding=True)
    translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

```

```python
# Batch translation function
def translate_sentences():
    while True:
        for i in range(10):
            sentence = input(f"\n[{i+1}/10] Enter an English sentence to translate into Spanish: ")
            spanish = translate(sentence)
            print(f"→ Spanish: {spanish}")

        again = input("\nDo you want to translate 10 more sentences? (yes/no): ").strip().lower()
        if again not in ['yes', 'y']:
            print("Finished translating. Goodbye!")
            break
```

```python
translate_sentences()
```
